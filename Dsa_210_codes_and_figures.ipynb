```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create DataFrame
data = {
    'Week': range(1, 53),
    'TV_Hours': [7.5, 5.3, 7.1, 5.5, 6.7, 5.2, 8.2, 7.7, 5.0, 5.0, 6.8, 4.1, 6.6, 5.2, 4.0,
                 3.1, 6.6, 0.0, 6.0, 3.1, 3.0, 5.5, 4.1, 2.8, 3.3, 3.1, 5.5, 5.1, 7.0, 0.0,
                 7.8, 9.3, 6.3, 10.5, 9.6, 4.0, 3.9, 3.7, 8.5, 0.0, 3.3, 2.5, 2.6, 2.3, 1.8,
                 1.6, 3.4, 3.7, 1.9, 0.0, 6.2, 3.7],
    'Gym_Days': [5, 5, 4, 5, 3, 2, 3, 3, 4, 3, 4, 3, 4, 3, 3, 2, 3, 5, 5, 5, 5, 5, 4, 3, 4,
                 3, 4, 3, 4, 3, 3, 2, 3, 3, 2, 3, 3, 2, 3, 3, 2, 2, 1, 1, 2, 3, 3, 3, 3, 3,
                 3, 3]
}

df = pd.DataFrame(data)

# Calculate correlation manually
def calculate_correlation(x, y):
    n = len(x)
    mean_x, mean_y = np.mean(x), np.mean(y)
    std_x, std_y = np.std(x), np.std(y)

    covariance = np.sum((x - mean_x) * (y - mean_y)) / (n - 1)
    correlation = covariance / (std_x * std_y)
    return correlation

correlation = calculate_correlation(df['TV_Hours'], df['Gym_Days'])

# Create a figure with multiple subplots
plt.style.use('default')
fig = plt.figure(figsize=(15, 10))

# 1. Time series plot
plt.subplot(2, 2, 1)
plt.plot(df['Week'], df['TV_Hours'], label='TV Hours', color='blue', marker='o', markersize=3)
plt.plot(df['Week'], df['Gym_Days'], label='Gym Days', color='red', marker='o', markersize=3)
plt.title('TV Hours and Gym Days Over Time')
plt.xlabel('Week')
plt.ylabel('Hours/Days')
plt.legend()
plt.grid(True)

# 2. Scatter plot with regression line
plt.subplot(2, 2, 2)
plt.scatter(df['TV_Hours'], df['Gym_Days'], alpha=0.5)

# Calculate and plot regression line
z = np.polyfit(df['TV_Hours'], df['Gym_Days'], 1)
p = np.poly1d(z)
x_range = np.linspace(df['TV_Hours'].min(), df['TV_Hours'].max(), 100)
plt.plot(x_range, p(x_range), 'r-', alpha=0.8)

plt.title(f'Correlation between TV Hours and Gym Days\nr={correlation:.2f}')
plt.xlabel('TV Hours per Week')
plt.ylabel('Gym Days per Week')
plt.grid(True)

# 3. Box plot
plt.subplot(2, 2, 3)
plt.boxplot([df['TV_Hours'], df['Gym_Days']], labels=['TV Hours', 'Gym Days'])
plt.title('Distribution of TV Hours and Gym Days')
plt.ylabel('Value')
plt.grid(True)

# 4. Histogram
plt.subplot(2, 2, 4)
plt.hist(df['TV_Hours'], bins=15, alpha=0.5, label='TV Hours', color='blue')
plt.hist(df['Gym_Days'], bins=15, alpha=0.5, label='Gym Days', color='red')
plt.title('Distribution of TV Hours and Gym Days')
plt.xlabel('Hours/Days')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

plt.tight_layout()

# Calculate and print summary statistics
summary_stats = df.describe()
print("\nSummary Statistics:")
print(summary_stats)

print("\nCorrelation Analysis:")
print(f"Correlation coefficient: {correlation:.3f}")
```
